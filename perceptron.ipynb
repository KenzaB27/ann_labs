```python
from data.binary_data import generate_data
from algorithms.perceptron import perceptron_learning
import matplotlib.pyplot as plt
import matplotlib
import numpy as np
from algorithms.delta import delta_rule
```

```python
n = 100
mA = [1.5, 2.0]
sigA = 0.5
mB = [-2.5, 0]
sigB = 0.5

X, y = generate_data(n, mA, sigA, mB, sigB, seed=42)
```

```python
w_per = perceptron_learning(X, y, lr=0.001, max_iters=250)
```

```python
w_del = delta_rule(X, y)
```

```python
colors = ['red','green']
plt.scatter(X[:,0], X[:,1], c=y, cmap=matplotlib.colors.ListedColormap(colors), s=10.0)

x1 = np.linspace(-2,2,100)
x2_per = (-w_per[0]*x1 -w_per[2]) / w_per[1]
x2_del = (-w_del[0]*x1 -w_del[2]) / w_del[1]

plt.plot(x1, x2_per, label='perceptron learning')
plt.plot(x1, x2_del, label='delta rule')
plt.legend()
plt.show()
```
