```python
from data.binary_data import generate_data
from algorithms.perceptron import perceptron_learning
import matplotlib.pyplot as plt
import matplotlib
import numpy as np
from algorithms.delta import delta_rule
import algorithms.plot as p
```

```python
n = 100
mA = [1.5, 2.0]
sigA = 0.5
mB = [-2.5, 0]
sigB = 0.5

X, y = generate_data(n, mA, sigA, mB, sigB, seed=39)
```

```python
w_per, history = perceptron_learning(X, y, lr=0.001, max_iters=250)
```

```python
p.plot_boundarys(history,X,y,'perceptron', 0.001)
```

```python
w_del, history = delta_rule(X, y,lr=0.1,sequential=1)
```

```python
p.plot_boundarys(history,X,y,'perceptron', 0.001)
```

```python
w_del, history = delta_rule(X, y,lr=0.1)
p.plot_boundarys(history,X,y,'perceptron', 0.001)
```

```python
colors = ['red','green']
plt.scatter(X[:,0], X[:,1], c=y, cmap=matplotlib.colors.ListedColormap(colors), s=2.0)

x1 = np.linspace(-2,2,100)
x2_per = (-w_per[0]*x1 -w_per[2]) / w_per[1]
x2_del = (-w_del[0]*x1 -w_del[2]) / w_del[1]

plt.plot(x1, x2_per, label='perceptron learning')
plt.plot(x1, x2_del, label='delta rule')
plt.legend()
plt.show()
```
