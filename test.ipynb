```python
import numpy as np
import matplotlib.pyplot as plt
import perceptron as ptn
plt.rcParams["figure.figsize"] = (10,6)
```

# Linearly separable data

```python
X1,y1 = ptn.create_training_data(1)
```

```python
bias = 1
eta = 0.1
neuron = ptn.Neuron(X1,y1,bias, eta)
```

```python
# training
for _ in range(20):
    neuron.update()
neuron.plot_boundarys()
```
